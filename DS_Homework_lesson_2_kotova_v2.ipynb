```python
import cianparser
import pandas as pd
```

```python
fpath = "C:\\Users\\123\\Documents\\ВШ\\DS\\"
fname=fpath + "cian_data_all_city.csv"
```

```python
list_city=['Москва','Санкт-Петербург','Екатеринбург','Новосибирск','Псков','Хабаровск','Томск','Иркутск','Владивосток','Магадан','Сочи','Петропавловск-Камчатский','Красноярск']
cnt_city=len(list_city)
print(cnt_city)
```

```python
i=0
cnt_row=20
while i < cnt_city:
    city_current = list_city[i]
    cian_parser = cianparser.CianParser (location=city_current)
    data = cian_parser.get_flats(deal_type="sale", rooms=("all"), with_saving_csv=False, additional_settings={"start_page":1, "end_page":cnt_row})
    df = pd.DataFrame(data)
    if i == 0:
        df.to_csv(fname,index= False )
    else:
        df.to_csv(fname, mode='a', index= False , header= False )
    i += 1
```
