Мастер, имея 10 деталей, из которых 3 – нестандартных, проверяет детали одну за другой, пока ему не попадется стандартная. Какова вероятность, что он проверит ровно две детали?

```python
# print(box)
n=3000000
i=1
cnt_true=0
while i<=n:
    box = ["nonstandart"]*3 + ['standart']*7
    cnt_true+= 1 if [box.pop(random.randint(0,len(box)-x)) for x in range(1,3)]==['nonstandart','standart'] else 0
    i+=1
p=round(cnt_true/n,5)
print(p)
```
